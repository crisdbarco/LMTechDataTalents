##### (a) A quantidade de abacates do tipo convecional e do tipo orgânico.

```python
print('Quantidade de abacates de cada tipo:\n', df.groupby('type').size())
```

##### (b) A quantidade de registros (linhas) de abacates vendidos para cada região.

```python
print('Quantidade de registros de vendas por região:\n', df.groupby('region').size())
```
